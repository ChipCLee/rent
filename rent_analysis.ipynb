```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('lujo_clean.csv', index_col='id')
print(type(df))
```

```python

df['time'] = pd.to_datetime(df['time'])
df['year'] = df['time'].dt.year
```

```python
import matplotlib as matplotlib
[f for f in matplotlib.font_manager.fontManager.ttflist]
```

```python
plt.rcParams['font.family']='Heiti TC'
plt.figure(figsize=(12,6))
```

```python
df_seleced=df.loc[(df['type']=='整層住家') | (df['type']=='雅房') | (df['type']=='分租套房') | (df['type']=='獨立套房') ]
sns.lineplot(x=df_seleced['area'],y=df_seleced['price'], hue=df_seleced['year'])
```

```python
sns.scatterplot(x=df_seleced['area'], y=df_seleced['price'], hue=df_seleced['type'])
```

```python
sns.lmplot(x="area", y="price", hue="type", data=df_seleced)
```

```python
sns.histplot(df_seleced['area'])
```

```python
sns.histplot(df_seleced['type'])
```

```python
sns.jointplot(x=df_seleced['area'], y=df_seleced['price'], kind="kde")
```
